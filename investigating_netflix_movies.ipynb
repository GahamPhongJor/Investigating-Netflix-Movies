## Investigating Netfix Movies  

```python
# importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read in the Netflix csv to a DataFrame
netflix_df = pd.read_csv('D:/Projects/netflix_data.csv')
netflix_df.shape
```

```python
# Subset the DataFrame for type "Movie"
netflix_movie = netflix_df[netflix_df['type'] == 'Movie']
netflix_movie.head(5)
```

```python
# Filter the to keep only movies released in the 1990s
netflix_1990s = netflix_movie[(netflix_movie['release_year'] >= 1990) & (netflix_movie['release_year'] <= 1999)]
netflix_1990s.head(5)
```

```python
# Visualize the duration column of your filtered data to see the distribution of movie durations
plt.hist(netflix_1990s['duration'])
plt.title('Distribution of Movie Durations in the 1990s')
plt.xlabel('Duration (Minutes)')
plt.ylabel('Number of Movies')
plt.show()
```

```python
# Filter the data again to keep only the Action movies
action_movie = netflix_1990s[netflix_1990s['genre'] == 'Action']
action_movie.head(5)
```

```python
# Count the value of action movies duration lessthan 90 minutes
no_of_action_movie_short = (action_movie['duration'] < 90).sum()
print(no_of_action_movie_short)
```

```python
# Another method to find the number of short action movies
# Use a for loop and a counter to count how many short action movies there were in the 1990s 

short_movie_count = 0

# Iterate over the labels and rows of the DataFrame and check if the duration is less than 90,
#  if it is, add 1 to the counter, if it isn't, the counter should remain the same

for index, row in action_movie.iterrows():
    if row['duration'] < 90:
        short_movie_count = short_movie_count + 1
    else:
        short_movie_count = short_movie_count

print(short_movie_count)
```
